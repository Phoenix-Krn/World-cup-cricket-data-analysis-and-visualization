````python
%matplotlib inline
import matplotlib.pyplot as plt #import the packages needed for visualization
import seaborn as sns
from matplotlib import image
````

````python
df1.drop("index",axis= 1,inplace=True)
df1.drop("Unnamed: 0",axis= 1,inplace=True)
df1
````

````python
top_performers=df1.sort_values(by="wickets",ascending=False)
top_performers
````

# **visualizing** the data using matplotlib.

````python
img=plt.imread("/content/bb2.jpg") #to import image for the background
fig=plt.figure(figsize=(10,10))
fig, ax=plt.subplots()
plt.style.use("dark_background")
plt.bar(top_performers["player"][0:12],top_performers["wickets"][0:12],color = ["red","b","grey","g","c","purple","aqua","pink","orange","white","yellow","brown"],width=0.5,align="center")
plt.xticks(rotation=90)
plt.xlabel("Player's name",c="blue")
plt.ylabel("wickets taken",c="blue")
plt.title(" Wickets taken by top 10 players")
ax.imshow(img, extent=[-1,11,0,8])
````

````python
pd.crosstab(df1["player"],df1["wickets"])
````

````python
fig=plt.figure(figsize=(15,55))
pd.crosstab(df1["player"],df1["wickets"]).head().plot(kind = "bar")
plt.xticks(rotation=90)
plt.xlabel("Teams",c="blue")
plt.ylabel("opponents",c="blue")
plt.title(" Teams and the opponents faced")
````

The above figure is a crosstab visulization

````python
img=plt.imread("/content/bb4.jpg")
fig=plt.figure(figsize=(50,80))
fig, ax=plt.subplots()
plt.scatter(x="runs",y="4s",data=df1,color="red",marker="*",label="4s")
plt.scatter(x="runs",y="6s",data=df1,color="blue",marker="*",label="6s")
plt.legend(title="scatterplot")
plt.xlabel("Runs",fontsize=15)
plt.ylabel("no. of 4's and 6's",fontsize=15)
ax.imshow(img,extent=[20,50,-1,10])
plt.show()
````

The above is a scatter plot representation of 4s and 6s

````python
#visulization using seaborn
sns.boxplot(x = "overs", y = "wickets", data = df1)
````

The above is a boxplot represenation
